```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python

# Group the data by job role and attrition, and calculate the average distance from home
grouped = df.groupby(['JobRole', 'Attrition'])['DistanceFromHome'].mean()
```

```python
# Create a bar chart to show the breakdown of distance from home by job role and attrition
ax = grouped.unstack().plot(kind='bar', figsize=(10, 6), width=0.8)
ax.set_ylabel('Average Distance From Home')
ax.set_xlabel('Job Role')
ax.set_title('Breakdown of Distance from Home by Job Role and Attrition')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Attrition', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Group the data by education and attrition, and calculate the average monthly income
grouped = df.groupby(['Education', 'Attrition'])['MonthlyIncome'].mean()
```

```python
# Create a bar chart to compare average monthly income by education and attrition
ax = grouped.unstack().plot(kind='bar', figsize=(10, 6), width=0.8)
ax.set_ylabel('Average Monthly Income')
ax.set_xlabel('Education Level')
ax.set_title('Comparison of Average Monthly Income by Education and Attrition')
plt.xticks(rotation=0, ha='center')
plt.legend(title='Attrition', bbox_to_anchor=(1, 1))
plt.show()
```
